```python
import os
import pandas as pd
```

```python
def read_txt(file_path):
    metrics = {}
    with open(file_path, 'r') as f:
        for line in f:
            if ':' in line:
                key, value = line.strip().split(':')
                metrics[key.strip()] = float(value.strip())
    return metrics
```
